```python
import pandas as pd
import requests
import plotly.express as px
import json
```

```python
def fetch_data():
    url = "https://web.archive.org/web/20240929211114/https://fruityvice.com/api/fruit/all"
    try:
        r = requests.get(url)
        r.raise_for_status() # for api working 
        data = r.json()
        #nested json into flat table
        df = pd.json_normalize(data)
        #Cleaning column names by removing the 'nutritions.' prefix
        df.columns = [c.replace("nutritions.","") for c in df.columns]
        return df
    except Exception as e:
        print(f"Error fetching data {e}")
        return None
    
df = fetch_data()
df.head()
```

```python
print(f"Total fruit analyzed {len(df)}")
print("\n--- Nutritional Summary ---")
display(df.describe())

```

```python
fig_scatter = px.scatter(df, 
                             x='sugar', 
                             y='carbohydrates', 
                             size='calories', 
                             color='family', 
                             hover_name='name',
                             title='Correlation: Sugar vs Carbohydrates')
fig_scatter.show()
```

```python
fig_bar =px.bar(df, 
                     x='name', 
                     y='calories', 
                     color='calories',
                     color_continuous_scale='Viridis', 
                     title='Ranking Fruits by Calorie Content (per 100g)',
                     labels={'name': 'Fruit Name', 'calories': 'Calories (kcal)'},
                     template='plotly_dark') 

    # Layout styling
fig_bar.update_layout(
        margin=dict(t=50, l=25, r=25, b=25),
        font=dict(family="Arial", size=12)
    )
fig_bar.show()
```

```python
fig_scatter = px.scatter(df, 
                             x='sugar', 
                             y='carbohydrates', 
                             size='calories', 
                             color='family', 
                             hover_name='name', 
                             title='Nutritional Correlation: Sugar vs. Total Carbohydrates',
                             labels={'sugar': 'Sugar (g)', 'carbohydrates': 'Total Carbs (g)'},
                             template='plotly_white')


fig_scatter.update_layout(
        margin=dict(t=50, l=25, r=25, b=25),
        font=dict(family="Arial", size=12)
    )
fig_scatter.show()
```
